```python
import cv2
import matplotlib.pyplot as plt
import pyautogui as robot
import numpy as np
```

```python
def imgshow(image):
    cv2.imshow('_', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def rect(face, image, color=(0,255,0)):
    for (x, y, w, h) in face:
        out = cv2.rectangle(image, (x, y), (w+x, h+y), color, 3)
    return out
```

```python
face_model = cv2.CascadeClassifier('aio_work/AI_tools/haarcascade_frontalface_default.xml')
```

### reading from camera

```python
cam = cv2.VideoCapture(0)
flag = True

while flag:
    _, img_cam = cam.read()
    img_cam = cv2.flip(img_cam, 1)
    gray = cv2.cvtColor(img_cam, cv2.COLOR_BGR2GRAY)

    face = face_model.detectMultiScale(gray)

    offset_zone = np.array([[190, 170, 270, 240]]) #the zone where you move your head but nothing happens
    # out = rect(face, img_cam.copy())
    out = rect(offset_zone, img_cam.copy() , (255,255,255))
    cv2.putText(out, 'put your face in the white box', (190,160), cv2.FONT_HERSHEY_PLAIN, 1, (255,255,255), 2)
    cv2.putText(out, 'move your face to move your cursor', (175,430), cv2.FONT_HERSHEY_PLAIN, 1, (255,255,255), 2)

    if (len(face) > 0):
        
        """changing the color when face gets out of the offset border"""
        for (x,y,w,h), (a,b,c,d) in zip(face, offset_zone):

            pos = robot.position()
            mouse_x = pos.x
            mouse_y = pos.y
            
            color = (0, 255, 0)

        
                
            if x<a:
                color = (180,12,90)
                cv2.putText(out, 'To the left', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                #in order to give it some momentum we do abs(x-a)
                mouse_x = mouse_x - abs(x - a)
                robot.moveTo(mouse_x, mouse_y)

            if (w+x > a+c):
                color = (180,12,90)
                cv2.putText(out, 'To the right', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                mouse_x = mouse_x + abs((w+x)-(a+c))
                robot.moveTo(mouse_x, mouse_y)

            if (y < b):
                color = (180,12,90)
                cv2.putText(out, 'To up', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                mouse_y = mouse_y - abs(y-b)
                robot.moveTo(mouse_x, mouse_y)

            if (h+y > d+b):
                color = (180,12,90)
                cv2.putText(out, 'To down', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                mouse_y = mouse_y + abs((h+y)-(d+b))
                robot.moveTo(mouse_x, mouse_y)

            
            
            out = rect(face, out, color)
            cv2.imshow('_', out)

        # for (x, y, w, h) in face:
        #     print(f'x: {x}, y: {y}')

            if cv2.waitKey(1) == ord('q'):
                flag = False
                cv2.destroyAllWindows()
                cam.release() #baraye ink durbin khamush she
                break
    
```
