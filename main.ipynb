```python
import cv2
import matplotlib.pyplot as plt
import pyautogui as robot
import numpy as np
```

```python
def imgshow(image):
    """ we use this function to show our video frames realtime"""
    cv2.imshow('_', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def rect(face, image, color=(0,255,0)):
    """ multiple rectangles are detected on your face, 
    this function gets the image and the coordinates, and draws the rectangle"""
    for (x, y, w, h) in face:
        out = cv2.rectangle(image, (x, y), (w+x, h+y), color, 3)
    return out
```

```python
face_model = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

### reading from camera and tracking your face!

```python
cam = cv2.VideoCapture(0) #0 refers to your laptop cam, if you got more cam, like webcam, make it 1
flag = True

while flag:
    _, img_cam = cam.read()
    img_cam = cv2.flip(img_cam, 1)
    gray = cv2.cvtColor(img_cam, cv2.COLOR_BGR2GRAY)

    face = face_model.detectMultiScale(gray)

    offset_zone = np.array([[190, 170, 270, 240]]) #the zone where you move your head but nothing happens (I call it safe zone!)
    
    out = rect(offset_zone, img_cam.copy() , (200,200,200))
    cv2.putText(out, 'put your face in the white box', (190,160), cv2.FONT_HERSHEY_PLAIN, 1, (200,200,200), 2)
    cv2.putText(out, 'move your face to move your cursor', (175,430), cv2.FONT_HERSHEY_PLAIN, 1, (200,200,200), 2)

    if (len(face) > 0):
        
        """changing the color when face gets out of the offset border"""
        for (x,y,w,h), (a,b,c,d) in zip(face, offset_zone):

            pos = robot.position()
            mouse_x = pos.x
            mouse_y = pos.y
            
            color = (0, 255, 0)

        
                
            if x<a:
                color = (180,12,90)
                cv2.putText(out, 'To the left', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                #in order to give it some momentum we do abs(x-a)
                mouse_x = mouse_x - abs(x - a)
                robot.moveTo(mouse_x, mouse_y)

            if (w+x > a+c):
                color = (180,12,90)
                cv2.putText(out, 'To the right', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                mouse_x = mouse_x + abs((w+x)-(a+c))
                robot.moveTo(mouse_x, mouse_y)

            if (y < b):
                color = (180,12,90)
                cv2.putText(out, 'To up', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                mouse_y = mouse_y - abs(y-b)
                robot.moveTo(mouse_x, mouse_y)

            if (h+y > d+b):
                color = (180,12,90)
                cv2.putText(out, 'To down', (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, color, 1)
                mouse_y = mouse_y + abs((h+y)-(d+b))
                robot.moveTo(mouse_x, mouse_y)

            
            
            out = rect(face, out, color)
            cv2.imshow('_', out)



            if cv2.waitKey(1) == ord('q'): #in order to get out of the camera, press 'Q'
                flag = False
                cv2.destroyAllWindows()
                cam.release() #to turn of your camera
                break
    
```
